```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('SBAnational.csv')

df=df.drop(columns=['LoanNr_ChkDgt','LowDoc','Name','City','State','Zip','Bank','BankState','ChgOffPrinGr','ApprovalDate','DisbursementDate','ApprovalFY','ChgOffDate','BalanceGross'],axis=1)

x=pd.get_dummies(df['RevLineCr'])
y=df['RevLineCr'].apply(lambda x:1 if x=='Y' else -1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
x_train. head()
```

```python
y_train. head()
```

```python
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score
```

```python
model = Sequential()
model.add(Dense(units=32, activation='relu', input_dim=len(x_train.columns)))
model.add(Dense(units=64, activation='relu'))
model.add(Dense(units=1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics='accuracy')
```

```python
model.fit(x_train, y_train, epochs=200, batch_size=32)
```

```python
y_hat = model.predict(x_test)
y_hat = [0 if val < 0.5 else 1 for val in y_hat]
```

```python
accuracy_score(y_test, y_hat)
```

```python
model.save('tfmodel')
```
